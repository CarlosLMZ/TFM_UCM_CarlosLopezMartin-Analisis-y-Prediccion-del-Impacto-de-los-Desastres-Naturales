```python
import pandas as pd
import numpy as np
```

```python
with open(r'C:\Users\Carlos\A-TFM\BancoMundial_PIB.csv', encoding='utf-8') as f:
    lines = f.readlines()

# Imprimir las primeras líneas para inspección
for i, line in enumerate(lines[:10]):
    print(f"Línea {i}: {line}")
```

```python
# Encontrar la fila con los encabezados reales
for i, line in enumerate(lines):
    if 'Country Name' in line and 'Indicator Name' in line:
        header_index = i
        break

# Leer el archivo desde esa fila en adelante
df = pd.read_csv(r'C:\Users\Carlos\A-TFM\BancoMundial_PIB.csv', skiprows=header_index)
```

```python
df.head(5)
```

```python
columnas_a_eliminar = ['Indicator Name', 'Indicator Code'] + [str(año) for año in range(1960, 1970)]

df_limpio = df.drop(columns=columnas_a_eliminar)
```

```python
df_limpio.head(5)
```

```python
columnas_a_eliminar2 = [str(año) for año in range(2022, 2025)]

if 'Unnamed: 69' in df_limpio.columns:
    columnas_a_eliminar2.append('Unnamed: 69')

df_limpio2 = df_limpio.drop(columns=columnas_a_eliminar2)
```

```python
df_limpio2.head(5)
```

```python
df_limpio2.shape
```

```python
df_limpio2.columns = df_limpio2.columns.astype(str)

columnas_a_limpiar = [col for col in df_limpio2.columns if col.isdigit() and 1970 <= int(col) <= 2021]

for col in columnas_a_limpiar:
    df_limpio2[col] = (
        df_limpio2[col].astype(str)
        .str.replace('.', '', regex=False)
        .replace('nan', pd.NA)
        .astype('Int64')
    )
```

```python
# Filtrar solo las columnas deseadas
df_filtrado = df_limpio2[['Country Name', 'Country Code', '2021']]
df_filtrado.head(5)
```

```python
df_limpio2.isnull().mean() * 100
```

```python
df_limpio3 = df_filtrado.dropna()
```

```python
df_limpio3.isnull().mean() * 100
```

```python
df_limpio3.rename(columns={'Country Code': 'ISO'}, inplace=True)
```

```python
df_limpio4 = df_limpio3.drop(columns=['Country Name'])
```

```python
df_limpio4.rename(columns={'2021': 'PIB_2021'}, inplace=True)
```

```python
df_limpio4.to_csv("PIB_2021.csv", index=False)
```
