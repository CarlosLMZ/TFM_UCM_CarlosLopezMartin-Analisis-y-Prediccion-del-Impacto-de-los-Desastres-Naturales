## Feature Engineering

```python
df_tram['Disaster_Subtype'].value_counts()
```

```python
threshold = 10

new_cat_subtypes = df_tram['Disaster_Subtype'].value_counts()
new_cat_subtypes = new_cat_subtypes[new_cat_subtypes <= threshold].index

df_tram['Disaster_Subtype'] = df_tram['Disaster_Subtype'].replace(new_cat_subtypes, 'Other')
```

```python
df_tram['Disaster_Subtype'].value_counts()
```

```python
# Glacial lake outburst y animal accident están muy poco representados
df_tram['Disaster_Type'].value_counts()
```

```python
# Cambiamos glacial lake outburst de categoría
df_tram['Disaster_Type'] = df_tram['Disaster_Type'].replace('Glacial lake outburst', 'Flood')
```

```python
# Eliminamos animal accident
df_tram = df_tram[df_tram['Disaster_Type'] != 'Animal accident']
```

```python
df_tram['Disaster_Subgroup'].value_counts()
```

```python
# Lectura del DataFrame creado del PIB por países en 2021
pd.set_option('display.max_rows', None, 'display.max_columns', None)
df_pib = pd.read_csv(r'C:\Users\Carlos\A-TFM\PIB_2021.csv')
```

```python
df_dis_pib = pd.merge(df_tram, df_pib, on='ISO', how='left')
```

```python
df_dis_pib.isnull().sum()
```

```python
df_dis_pib[df_dis_pib['PIB_2021'].isna()]['ISO'].unique()
```

```python
df_dis_pib['PIB_2021'].describe()
```

```python
df_dis_pib['PIB_2021'] = df_dis_pib['PIB_2021'].fillna('Unknown')
```

```python
# Filtramos para dejar a un lado los valores desconocidos
pib_group = df_dis_pib['PIB_2021'] != 'Unknown'

# Dividimos por cuartiles y asignamos etiqueta en una nueva columna
df_dis_pib.loc[pib_group, 'Grupo_Riqueza'] = pd.qcut(df_dis_pib.loc[pib_group, 'PIB_2021'].astype(float), q=4,
                                        labels=['Low', 'MidLow', 'MidHigh', 'High'])

# Pasamos a categórica
df_dis_pib['Grupo_Riqueza'] = df_dis_pib['Grupo_Riqueza'].astype('category')

# Agregamos 'Unknown' como categoría válida
if 'Unknown' not in df_dis_pib['Grupo_Riqueza'].cat.categories:
    df_dis_pib['Grupo_Riqueza'] = df_dis_pib['Grupo_Riqueza'].cat.add_categories(['Unknown'])
    
# Metemos de nuevo los valores desconocidos
df_dis_pib.loc[~pib_group, 'Grupo_Riqueza'] = 'Unknown'
```

```python
df_dis_pib.head(5)
```

```python
df_dis_pib['Grupo_Riqueza'].value_counts()
```

```python
# Retomamos la variable objetivo
# Lista muy larga, mucha concentración de datos en pocas muertes

#df_dis_pib['Total_Deaths'].value_counts()
```

```python
# Vamos a crear dos categorías, con un umbral de 100 muertes para diferenciar.
df_dis_pib['Impact'] = df_dis_pib['Total_Deaths'].apply(lambda x: 1 if x > 100 else 0)
```

# Modelos Machine Learning

```python
# Seleccionamos las variables que hemos preparado anteriormente
variables_modelo = [
    'Year',
    'Disaster_Subgroup',
    'Disaster_Type',
    'Disaster_Subtype',
    'Region',
    'Continent',
    'Dis_Mag_Scale',
    'Duration_Days',
    'Season',
    'Grupo_Riqueza',
    'Impact'
]
```

```python
df_model = df_limpio[variables_modelo].copy()
```

```python
df_model.info()
```

```python
cat_cols = [
    'Year',
    'Disaster_Subgroup',
    'Disaster_Type',
    'Disaster_Subtype',
    'Region',
    'Continent',
    'Dis_Mag_Scale',
    'Duration_Days',
    'Season',
    'Grupo_Riqueza',
    'Impact'
]

df_model[cat_cols] = df_model[cat_cols].astype('category')
```

```python
# Confirmamos que todos los tipos de variables sean correctos
df_model.info()
```

```python
# Preparamos X e y
X = df_model.drop(columns='Impact')
y = df_model['Impact']
```

```python
label_encoders = {}
for col in X.columns:
    if X[col].dtype.name == 'category':
        le = LabelEncoder()
        X[col] = le.fit_transform(X[col])
        label_encoders[col] = le
```

```python
# Dividimos en train y test
# Probamos con stratify por el desbalanceo que tenemos
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=2025, stratify=y
)
```

## Hyperparameter tuning (preparación)

```python
rf = RandomForestClassifier(random_state=11)

# Definimos los hiperparámetros
params_GS = {
    'n_estimators': [100, 200, 300, 400, 500],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10, 15],
    'min_samples_leaf': [1, 2, 4, 6],
    'bootstrap': [True, False]
}

grid_search = GridSearchCV(
    estimator=rf,
    param_grid=params_GS,
    cv=10,
    scoring='f1_weighted',
    n_jobs=-1,
    verbose=2
)
```

```python
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
params_GS2 = {
    'n_estimators': [100, 200, 300, 400, 500],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10, 15],
    'min_samples_leaf': [1, 2, 4, 6],
    'bootstrap': [True, False]
}

grid_search2 = GridSearchCV(
    estimator=rf,
    param_grid=params_GS2,
    cv=10,
    scoring='roc_auc',
    n_jobs=-1,
    verbose=2
)

grid_search2.fit(X_train, y_train)
```

```python
grid_search2.best_params_
```

```python
# Vamos a probar a ajustar los pesos por clase dado el enorme desbalance que tenemos de clases en la var. obj.
class_weights = compute_class_weight(class_weight='balanced', classes=np.unique(y_train), y=y_train)

weights_dict = dict(zip(np.unique(y_train), class_weights))
```

```python
# Definimos el modelo con los best_params. 
# Usamos los que nos dió GridSearch con roc_auc.

model = RandomForestClassifier(
    n_estimators=500,
    max_depth=10,
    min_samples_split=15,
    min_samples_leaf=2,
    max_features='sqrt',
    bootstrap=True,
    class_weight=weights_dict,
    random_state=16,
    n_jobs=-1
)
```

```python
# Entrenamos el modelo
model.fit(X_train, y_train)
```

```python
# Predecimos
y_pred = model.predict(X_test)
```

```python
# Predecimos las probabilidades de caer en cada clase
# Que nos muestre un vector en vez de una matriz para evitar errores con el score roc_auc
y_proba = model.predict_proba(X_test)[:, 1]
```

```python
print("ROC AUC:", roc_auc_score(y_test, y_proba))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
# Graficamos la matriz de confusión
conf_mat = confusion_matrix(y_test, y_pred)

conf_mat_display = ConfusionMatrixDisplay(conf_mat).plot()
conf_mat_display
plt.show()
```

```python
# Graficamos el área de la curva ROC
fpr, tpr, thresholds = roc_curve(y_test, y_proba)

roc_auc = roc_auc_score(y_test, y_proba)

# Graficar
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random classifier')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Tasa de falsos positivos (FPR)')
plt.ylabel('Tasa de verdaderos positivos (TPR)')
plt.title('Curva ROC')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

```python
# Graficamos el report
report_dict = classification_report(y_test, y_pred, output_dict=True)

df_report = pd.DataFrame(report_dict).transpose()

df_classes = df_report.loc[['0', '1'], ['precision', 'recall', 'f1-score']]

df_classes.plot(kind='bar', figsize=(8, 6), colormap='cool')
plt.title('Métricas por clase')
plt.ylabel('Valor')
plt.ylim(0, 1.05)
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.legend(loc='lower right')
plt.tight_layout()
plt.show()
```

```python
# Sacamos el Feature Importance
importances = model.feature_importances_
feature_names = X_train.columns

feat_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
feat_df = feat_df.sort_values(by='Importance', ascending=False)

feat_df
```

```python
plt.figure(figsize=(10, 6))
plt.barh(feat_df['Feature'], feat_df['Importance'], color='steelblue')
plt.gca().invert_yaxis()
plt.title('Importancia de las Variables')
plt.xlabel('Importancia')
plt.tight_layout()
plt.show()
```

```python
# Tratamos de ver un mejor punto de corte o punto de corte óptimo
y_probs = model.predict_proba(X_test)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, y_probs)

youden_ind = tpr - fpr
best_idx = youden_ind.argmax()
optimal_threshold = thresholds[best_idx]

print(f"Umbral óptimo: {optimal_threshold:.2f}")
```

```python
thresholds = np.arange(0.0, 1.0, 0.01)
f1_scores = [f1_score(y_test, y_probs > t) for t in thresholds]

best_threshold = thresholds[np.argmax(f1_scores)]

print(f"Umbral que maximiza F1: {best_threshold:.2f}")
```

```python
# Ajustamos a ver cómo queda
threshold = 0.38
y_pred_youden = (y_probs >= threshold).astype(int)

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_youden))

print("\nClassification Report:")
print(classification_report(y_test, y_pred_youden))

print("F1 Score:", f1_score(y_test, y_pred_youden))
```

```python
# Definimos función para sacar el rendimiento del F1 de los modelos
def plot_learning_curve(model, name, X, y, ax, cv=10, scoring='f1'):
    train_sizes, train_scores, val_scores = learning_curve(
        model, X, y, cv=cv, scoring=scoring, n_jobs=-1,
        train_sizes=np.linspace(0.1, 1.0, 10), shuffle=True, random_state=11
    )
    
    train_mean = np.mean(train_scores, axis=1)
    val_mean = np.mean(val_scores, axis=1)
    
    ax.plot(train_sizes, train_mean, label='Train', color='blue')
    ax.plot(train_sizes, val_mean, label='Validation', color='orange')
    ax.set_title(name, fontsize=16)
    ax.set_xlabel('Tamaño del conjunto de entrenamiento')
    ax.set_ylabel(scoring)
    ax.set_ylim(0, 1)
    ax.grid(True)
    ax.legend()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 7))

plot_learning_curve(model_red, 'Random Forest', X_train_red, y_train_red, axes[0])
plot_learning_curve(modelo_xgb_red, 'XGBoost', X_train_red, y_train_red, axes[1])
plot_learning_curve(model_CB, 'CatBoost', X_train_red, y_train_red, axes[2])

plt.suptitle('Curva de Aprendizaje', fontsize=18)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
